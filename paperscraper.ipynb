```python
import os
import os
import requests
from urllib.parse import urljoin
from bs4 import BeautifulSoup
import pandas as pd




def extract_url_pdf(input_url, AL=True):

    months = ['june', 'march', 'november']
    years = ['2022', '2023', '2024', '2025']
    filterwords = ['count', 'speaking', 'endorsement', 'speaking', 'coursework', 'double', 'award', 
                'component', 'paper', 'practical', 'oral', 'performance', 'written', 'without', 'qualification']
    url = input_url

    #If there is no such folder, the script will create one automatically

    response = requests.get(url)
    soup= BeautifulSoup(response.text, "html.parser") 

    for link in soup.select("a[href$='.pdf']"):
        #Name the pdf files using the last portion of each link which are unique in this case
        filename = link['href'].split('/')[-1][7:]
        sortname = (filename.replace('cambridge-','')
                    .replace('theshold','threshold')
                    .replace('igcse-','')
                    .replace('9-1','91')
                    .replace('-grade-threshold-table','')
                    .replace('-grade-thresholds','')
                    .replace('-grade-threshold','')
                    )
        filespecifics = sortname.replace('.pdf','').split('-')
        filespecifics = [i for i in filespecifics if i not in months 
                 and i not in years
                 and not i.isdigit()
                 and i != ''
                 and i != ' '
                 and i not in filterwords
                 or i == '91'
                 ]
        
        
        #print(filespecifics)
        
        if '91' in filespecifics:
            filespecifics = [i for i in filespecifics if i != '91']
            filespecifics.append('9-1')
            filespecifics = ['-'.join(filespecifics)]
            filespecifics.append('9-1')
        else:
            filespecifics = ['-'.join(filespecifics)]
        
        filespecifics[0] = (filespecifics[0].replace('english-literature', 'literature')
                             .replace('sciences-co-ordinated','science-co-ordinated')
                             .replace('communications','communication')
                             .replace('mathematics-additional','additional-mathematics')
                             .replace('travel-tourism','travel-and-tourism')
                             .replace('art-design','art-and-design')
                             .replace('english-as-a-second-language-in','english-as-a-second-language')
                             .replace('french-first-language','first-language-french')
                             .replace('afrikaans-second-langauage','afrikaans-second-language')
                             .replace('techology','technology')
                             .replace('spanish-first-language','first-language-spanish')
                             .replace('turkish-first-language','first-language-turkish')
                             .replace('thai-first-language','first-language-thai')
                             .replace('vietnamese-first-language','first-language-vietnamese')
                             .replace('portuguese-first-language','first-language-portuguese')
                             .replace('malay-first-language','first-language-malay')
                             .replace('german-first-language','first-language-german')
                             .replace('food-nutrition','food-and-nutrition')
                             .replace('english-first-language','first-language-english')
                             .replace('english-first-language-us','first-language-english-us')
                             .replace('design-technology','design-and-technology')
                             .replace('economic','economics')
                             .replace('chinese-first-language','first-language-chinese')
                             .replace('business-studies','business')
                             .replace('arabic-first-language','first-language-arabic')
                             .replace('economicss','economics')
                             .replace('english-second-language','english-as-a-second-language')
                             .replace('first-language-spanish','spanish')
                             .replace('spanish','first-language-spanish')
                             .replace('chinese-second-language','chinese-as-a-second-language')
                             .replace('global-perspectives-research','global-perspectives-and-research')
                             .replace('econimics','economics')
                             .replace('-a-level','')
                             .replace('a-level-','')
                             )
        
        #print(filespecifics)
        
        foldername  = filespecifics[0]
        
        if AL:
            filename = './thresholds/AL/' + foldername + '/' + filename
            folderpath = './thresholds/AL/' + foldername + '/'
            if not os.path.exists(folderpath): os.mkdir(folderpath)
        else:
            filename = './thresholds/IGCSE/' + foldername + '/' + filename.replace('91','9-1')
            folderpath = './thresholds/IGCSE/' + foldername + '/'
            if not os.path.exists(folderpath): os.mkdir(folderpath)
        print(filename)
        with open(filename, 'wb') as f:
            print(filename)
            f.write(requests.get(urljoin(url,link['href'])).content)


years_AL = [
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/november-2025/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/june-2025/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/march-2025/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/november-2024/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/june-2024/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/march-2024/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/november-2023/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/june-2023/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/march-2023/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/november-2022/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/june-2022/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-advanced/cambridge-international-as-and-a-levels/grade-threshold-tables/march-2022/index.aspx",
]

years_IGCSE = [
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/june-2025/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/march-2025/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/november-2024/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/june-2024/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/march-2024/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/november-2023/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/june-2023/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/march-2023/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/november-2022/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/june-2022/index.aspx",
    "https://www.cambridgeinternational.org/programmes-and-qualifications/cambridge-upper-secondary/cambridge-igcse/grade-threshold-tables/march-2022/index.aspx",
]

for year in years_AL:
     extract_url_pdf(year, True)

for year in years_IGCSE:
    extract_url_pdf(year, False)
```
