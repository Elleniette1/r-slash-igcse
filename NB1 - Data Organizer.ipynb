```python
import paperparser
import os
import pandas as pd
```

```python
bigdf = pd.DataFrame()
reada = os.listdir("thresholds/AL")
reada = [i for i in reada if not i.endswith(".xlsx")]
for foldername in reada:
    #print(foldername)
    df = pd.DataFrame()
    readfiles = os.listdir(f"thresholds/AL/{foldername}")
    readfiles = [i for i in readfiles if not i.endswith(".xlsx")]
    for filename in readfiles:
        #print(filename)
        df = pd.concat([df, paperparser.parse_threshold(f"./thresholds/AL/{foldername}/" + filename)[2]])
        df.to_excel(f"./thresholds/AL/{foldername}/{foldername}-thresholds.xlsx")
    bigdf["subject"] = f"{foldername}"
    bigdf = pd.concat([bigdf, df])
    #display(df)
display(bigdf)
```

```python
bigdf.to_excel("./thresholds/AL/al-thresholds.xlsx")
```
